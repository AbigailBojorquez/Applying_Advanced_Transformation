```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
info_df = pd.read_csv('Data/superhero_info - superhero_info.csv')
info_df.head()
```

```python
info_df.info()
```

```python
sup_pow_df = pd.read_csv('Data/superhero_powers - superhero_powers.csv')
sup_pow_df.head()
```

```python
sup_pow_df.info()
```

```python
#name and publisher
info_df['Hero|Publisher'].head()
```

```python
## adding expand=True
info_df[['Hero Name', 'Publisher']] = info_df\
    ['Hero|Publisher'].str.split('|',expand=True)
info_df = info_df.drop(columns = ['Hero|Publisher'])
info_df.head(3)
```

```python
print(type(info_df.loc[0, 'Measurements']))
info_df.loc[0, 'Measurements']
```

```python
info_df['Measurements'] = info_df['Measurements'].str\
            .replace("'", '"')
info_df['Measurements'] = info_df['Measurements']\
            .apply(json.loads)
info_df.head(3)
```

```python
print(type(info_df.loc[0, 'Measurements']))
info_df.loc[0, 'Measurements']
```

```python
measurements = info_df['Measurements'].apply(pd.Series)
measurements
```

```python
info_df = pd.concat((info_df, measurements), axis = 1)
info_df = info_df.drop(columns = ['Measurements'])
```

```python
to_replace = [' cm',' kg']

for char in to_replace:
    info_df['Height'] = info_df['Height']\
                    .str.replace(char, '', regex=False)
    info_df['Weight'] = info_df['Weight']\
                    .str.replace(char, '', regex=False)

columns = ['Height', 'Weight']
for col in columns:
    info_df[col] = pd.to_numeric(info_df[col])
```

```python
info_df = info_df.reindex(columns=[\
            'Hero Name', 'Publisher', 'Gender', 'Eye color',\
            'Race', 'Hair color', 'Height', 'Skin color',\
            'Alignment', 'Weight'])
info_df.head(3)
```

```python
#Combining DF
hero_pow = sup_pow_df['hero_names']\
            .isin(info_df['Hero Name'])

sup_pow_df = sup_pow_df[hero_pow]
sup_pow_df.info()
```

```python
sup_pow_df = sup_pow_df.rename(columns =\
                                    {'hero_names':'Hero Name'})
```

```python
merg_sup_df = pd.merge(info_df, sup_pow_df,\
                          on='Hero Name', how='inner')
merg_sup_df.head(3)
```

```python
#one-hot-encoded columns for every power that appears in the dataset
print(type(merg_sup_df.loc[0,'Powers']))
merg_sup_df.loc[0,'Powers']
```

```python
spl_pow_df = merg_sup_df['Powers'].str.split(",")
```

```python
exp= spl_pow_df.explode('Powers')
exp.head(3)
```

```python
#saving unique values from exploded column
colum = exp.dropna().unique()
colum
```

```python
for col in colum:
    merg_sup_df[col] = merg_sup_df['Powers'].str.contains(col)
merg_sup_df.head()
```
